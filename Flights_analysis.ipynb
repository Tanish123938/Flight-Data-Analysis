#### Q.2. Show Bar Graphs representing the Departure Time & Arrival Time.

```python
data.head()
```

```python
data["departure_time"].value_counts()
```

```python
data["arrival_time"].value_counts()
```

```python
plt.figure(figsize=(16,4))

plt.subplot(1,2,1)

plt.bar(data["departure_time"].value_counts().index ,data["departure_time"].value_counts().values, color=["orange","green"])
plt.title("Departure Time")
plt.xlabel("D. Time")
plt.ylabel("D. Freq")

plt.subplot(1,2,2)
plt.bar(data["arrival_time"].value_counts().index ,data["arrival_time"].value_counts().values, color=["orange","green"])
plt.title("Arrival Time")
plt.xlabel("A. Time")
plt.ylabel("A. Freq")

plt.show()
```

#### Q.3. Show Bar Graphs representing the Source City & Destination City.

```python
data.head()
```

```python
data["source_city"].value_counts()
```

```python
data["destination_city"].value_counts()
```

```python
plt.figure(figsize=(16,4))

plt.subplot(1,2,1)

plt.barh(data["source_city"].value_counts().index ,data["source_city"].value_counts().values, color=["orange","green"])
plt.title("Source City")
plt.xlabel("Cities")
plt.ylabel("No. of flights")

plt.subplot(1,2,2)

plt.barh(data["destination_city"].value_counts().index ,data["destination_city"].value_counts().values, color=["orange","red"])
plt.title("Destination City")
plt.xlabel("Cities")
plt.ylabel("No. of flights")

plt.show()
```
